```python
import warnings
import matplotlib as mpl
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning, module="matplotlib")

# 한글 폰트 설정 (macOS 기준)
mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 0. 설정
file_path = "../연관분석/입주공장_면적x평당매매가추가_output.xlsx"

# 1. 데이터 로드 및 정제
df = pd.read_excel(file_path, engine="openpyxl")

# 수치형 컬럼만 보기 좋게(크게) 출력
num_df = df.select_dtypes(include=[np.number])
corr_matrix = num_df.corr()

print("■ 전체 수치형 변수 상관관계 행렬 (피어슨)")
with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.width', 200, 'display.precision', 2):
    print(corr_matrix)

# 시각화 (heatmap)
import seaborn as sns
plt.figure(figsize=(max(12, len(num_df.columns)), max(10, len(num_df.columns)//2)))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("상관관계 행렬(피어슨)")
plt.tight_layout()
plt.show()
```
