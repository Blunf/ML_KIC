```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET

def search_patents_by_applicant(applicant, service_key, num_of_rows=100, page_no=1):
    url = "http://plus.kipris.or.kr/kipo-api/kipi/patUtiModInfoSearchSevice/getAdvancedSearch"
    params = {
        'ServiceKey': service_key,
        'applicant': applicant,
        'numOfRows': num_of_rows,
        'pageNo': page_no
    }
    response = requests.get(url, params=params)
    results = []
    if response.status_code == 200:
        try:
            root = ET.fromstring(response.content)
        except ET.ParseError as e:
            print("XML 파싱 오류:", e)
            return []
        items = root.findall('.//item')
        for item in items:
            title = item.findtext('inventionTitle')
            abstract = item.findtext('astrtCont')
            results.append({
                "회사명": applicant,
                "발명의 명칭": title,
                "초록": abstract
            })
        return results
    else:
        print(f"API 요청 실패: {response.status_code}")
        return []

df_comp = pd.read_csv("../지산데이터전처리/지산내_입주업체/전국지식산업센터_입주업체.csv", encoding="utf-8")
df_comp = df_comp.iloc[1:100]  # 1행부터 15행까지만 남기기
company_names = df_comp.iloc[:, 1].dropna().unique()  # 2열(인덱스 1)에서 중복/결측 제거

service_key = "rE0GweYTXSE4DUHAj8tkC2gbtCktVqS22eAlxXr9fF0="
all_results = []

for company in company_names:
    print(f"크롤링 중: {company}")
    results = search_patents_by_applicant(company, service_key)
    all_results.extend(results)

# DataFrame으로 저장 및 CSV 추출
df_patents = pd.DataFrame(all_results)
df_patents.to_csv("모든기업_특허목록.csv", index=False)
print(df_patents)
```
