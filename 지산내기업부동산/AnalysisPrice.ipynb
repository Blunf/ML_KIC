```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 데이터 로드
data = pd.read_csv('KIC_price_4.csv')

# "강원특별자치도 원주시 반곡동"으로 시작하는 지번주소 필터링
filtered_data = data[data['지번주소'].str.startswith('경기 고양시 덕양구')]

# 필요한 열 선택 (전용면적과 월세)
filtered_data = filtered_data[['전용면적', '월세']].dropna()

# X(전용면적)과 y(월세) 정의
X = filtered_data[['전용면적']]  # 독립 변수
y = filtered_data['월세']        # 종속 변수

# 선형 회귀 모델 생성 및 학습
model = LinearRegression()
model.fit(X, y)

# 회귀 계수 및 절편 출력
print(f"회귀 계수(기울기): {model.coef_[0]}")
print(f"절편: {model.intercept_}")

# 예측값 계산
y_pred = model.predict(X)

# R^2 점수 출력 (모델의 설명력)
r2 = r2_score(y, y_pred)
print(f"R^2 점수: {r2}")

# 데이터 시각화
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='실제 데이터')  # 실제 데이터
plt.plot(X, y_pred, color='red', label='회귀선')     # 회귀선
plt.xlabel('전용면적 (㎡)', fontsize=12)
plt.ylabel('월세 (만원)', fontsize=12)
plt.title('경기 고양시 덕양구', fontsize=15)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import folium
from folium.plugins import MarkerCluster

# 데이터 로드
data = pd.read_csv('KIC_price_4.csv')

# 위도와 경도 데이터 추출
location_data = data[['위도', '경도']].dropna()

# 데이터 표준화
scaler = StandardScaler()
scaled_locations = scaler.fit_transform(location_data)

# DBSCAN 클러스터링
dbscan = DBSCAN(eps=0.5, min_samples=5)  # eps: 거리 임계값, min_samples: 최소 샘플 수
location_data['클러스터'] = dbscan.fit_predict(scaled_locations)

# 클러스터링 결과 확인
print(location_data['클러스터'].value_counts())

# 지도 생성
map = folium.Map(location=[location_data['위도'].mean(), location_data['경도'].mean()], zoom_start=12)

# 클러스터별로 마커 추가
marker_cluster = MarkerCluster()
for idx, row in location_data.iterrows():
    cluster = row['클러스터']
    if cluster != -1:  # -1은 노이즈 데이터
        folium.Marker(
            location=[row['위도'], row['경도']],
            popup=f"클러스터: {cluster}",
            icon=folium.Icon(color='blue' if cluster == 0 else 'green')
        ).add_to(marker_cluster)

marker_cluster.add_to(map)

# 지도 저장 및 출력
map.save('clustered_map.html')
print("클러스터링 결과를 지도에 저장했습니다: clustered_map.html")
```
