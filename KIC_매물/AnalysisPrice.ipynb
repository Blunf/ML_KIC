```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import folium
from folium.plugins import MarkerCluster

# 데이터 로드
data = pd.read_csv('KIC_price_4.csv')

# 위도와 경도 데이터 추출
location_data = data[['위도', '경도']].dropna()

# 데이터 표준화
scaler = StandardScaler()
scaled_locations = scaler.fit_transform(location_data)

# DBSCAN 클러스터링
dbscan = DBSCAN(eps=0.5, min_samples=5)  # eps: 거리 임계값, min_samples: 최소 샘플 수
location_data['클러스터'] = dbscan.fit_predict(scaled_locations)

# 클러스터링 결과 확인
print(location_data['클러스터'].value_counts())

# 지도 생성
map = folium.Map(location=[location_data['위도'].mean(), location_data['경도'].mean()], zoom_start=12)

# 클러스터별로 마커 추가
marker_cluster = MarkerCluster()
for idx, row in location_data.iterrows():
    cluster = row['클러스터']
    if cluster != -1:  # -1은 노이즈 데이터
        folium.Marker(
            location=[row['위도'], row['경도']],
            popup=f"클러스터: {cluster}",
            icon=folium.Icon(color='blue' if cluster == 0 else 'green')
        ).add_to(marker_cluster)

marker_cluster.add_to(map)

# 지도 저장 및 출력
map.save('clustered_map.html')
print("클러스터링 결과를 지도에 저장했습니다: clustered_map.html")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 데이터 로드
data = pd.read_csv('KIC_price_4.csv')

# 지번주소에서 동 추출
data['동'] = data['지번주소'].str.split(' ').str[:3].str.join(' ')  # 예: "경기 군포시 당정동"

# 결과를 저장할 리스트 초기화
results = []

# 동별로 그룹화하여 선형회귀 수행
for dong, group in data.groupby('동'):
    # 필요한 열 선택 (전용면적과 월세)
    filtered_data = group[['전용면적', '월세']].dropna()
    
    # 데이터가 충분하지 않으면 건너뜀
    if len(filtered_data) < 2:
        print(f"{dong} 데이터가 부족하여 건너뜁니다.")
        continue
    
    # X(전용면적)과 y(월세) 정의
    X = filtered_data[['전용면적']]  # 독립 변수
    y = filtered_data['월세']        # 종속 변수
    
    # 선형 회귀 모델 생성 및 학습
    model = LinearRegression()
    model.fit(X, y)
    
    # R^2 점수 계산
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    
    # 결과 저장
    results.append({'동': dong, 'R^2': r2})
    print(f"{dong} 처리 완료 - R^2: {r2:.2f}")

# 결과를 데이터프레임으로 변환
results_df = pd.DataFrame(results)

# CSV 파일로 저장
results_df.to_csv('RegressionResult.csv', index=False, encoding='utf-8-sig')
print("동별 R^2 값을 저장했습니다.")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np

# 데이터 로드
data = pd.read_csv('KIC_price_4.csv')

# 지번주소에서 시, 구, 동 추출
data['시'] = data['지번주소'].str.split(' ').str[1]  # 예: "경기 성남시"에서 "성남시"
data['구'] = data['지번주소'].str.split(' ').str[:3].str.join(' ')  # 예: "경기 성남시 분당구"
data['동'] = data['지번주소'].str.split(' ').str[:4].str.join(' ')  # 예: "경기 성남시 분당구 정자동"

# 결과를 저장할 리스트 초기화
final_results = []

# 1. 시별로 그룹화하여 선형회귀 수행
for city, group in data.groupby('시'):
    filtered_data = group[['전용면적', '월세']].dropna()
    if len(filtered_data) >= 2:
        X = filtered_data[['전용면적']]
        y = filtered_data['월세']
        model = LinearRegression()
        model.fit(X, y)
        r2 = r2_score(y, model.predict(X))
        if r2 >= 0.8:
            final_results.append({'군집': city, 'R^2': r2, '단계': '시'})
            continue

    # 2. 구별로 그룹화하여 선형회귀 수행
    for district, sub_group in group.groupby('구'):
        filtered_data = sub_group[['전용면적', '월세']].dropna()
        if len(filtered_data) >= 2:
            X = filtered_data[['전용면적']]
            y = filtered_data['월세']
            model = LinearRegression()
            model.fit(X, y)
            r2 = r2_score(y, model.predict(X))
            if r2 >= 0.8:
                final_results.append({'군집': district, 'R^2': r2, '단계': '구'})
                continue

        # 3. 동별로 그룹화하여 선형회귀 수행
        for dong, sub_sub_group in sub_group.groupby('동'):
            filtered_data = sub_sub_group[['전용면적', '월세']].dropna()
            if len(filtered_data) >= 2:
                X = filtered_data[['전용면적']]
                y = filtered_data['월세']
                model = LinearRegression()
                model.fit(X, y)
                r2 = r2_score(y, model.predict(X))
                if r2 >= 0.8:
                    final_results.append({'군집': dong, 'R^2': r2, '단계': '동'})
                    continue

            # 4. 동 내부에서 평당 월세를 기반으로 클러스터링 수행
            clustering_data = sub_sub_group[['평당월세']].dropna()

            # 데이터 정제: NaN, 무한 값, 너무 큰 값 제거
            clustering_data = clustering_data.replace([np.inf, -np.inf], np.nan).dropna()
            clustering_data = clustering_data[(clustering_data['평당월세'] < 1e6)]  # 비정상적으로 큰 값 제거

            if len(clustering_data) >= 2:
                # 데이터 표준화
                scaler = StandardScaler()
                scaled_data = scaler.fit_transform(clustering_data)

                # DBSCAN 클러스터링 수행
                dbscan = DBSCAN(eps=0.5, min_samples=5)  # eps: 거리 임계값, min_samples: 최소 샘플 수
                clustering_data['군집번호'] = dbscan.fit_predict(scaled_data)

                # 클러스터링 결과 저장
                for cluster_id in clustering_data['군집번호'].unique():
                    final_results.append({'군집': f"{dong}_클러스터_{cluster_id}", 'R^2': None, '단계': '클러스터', '클러스터 번호': cluster_id})
            else:
                # 샘플 부족 시 -1 클러스터로 처리
                final_results.append({'군집': dong, 'R^2': None, '단계': '샘플 부족', '클러스터 번호': -1})
                print(f"{dong} 클러스터링을 위한 샘플 수 부족: {len(clustering_data)}")

# 결과를 데이터프레임으로 변환
final_results_df = pd.DataFrame(final_results)

# CSV 파일로 저장
final_results_df.to_csv('Final_Clustered_Results_DBSCAN.csv', index=False, encoding='utf-8-sig')
print("최종 군집 결과를 Final_Clustered_Results_DBSCAN.csv로 저장했습니다.")
```

클러스터별 가격동향분석

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 데이터 로드
df = pd.read_csv("KIC_price.csv")

results = []

for cluster, group in df.groupby('클러스터'):
    filtered = group[['전용면적', '월세']].dropna()
    # 데이터가 2개 이상일 때만 회귀분석
    if len(filtered) < 2:
        continue
    X = filtered[['전용면적']]
    y = filtered['월세']
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    results.append({'클러스터': cluster, 'R2': r2})

# 결과 저장
results_df = pd.DataFrame(results)
results_df.to_csv("클러스터별회귀분석.csv", index=False, encoding='utf-8-sig')
print("클러스터별 R² 결과를 클러스터별회귀분석.csv로 저장했습니다.")
```
