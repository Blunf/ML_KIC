```python
import requests
import pandas as pd
import time

# 📌 카카오 REST API 키 설정
KAKAO_API_KEY = "f32e9311c06e941799af2d38345741de"
HEADERS = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}

# 📌 검색 키워드와 카테고리 코드
KEYWORD = "국민은행"
CATEGORY_GROUP_CODE = "BK9"  # 은행 카테고리

# 📌 검색할 지역 목록
REGIONS = ["서울", "부산", "대구", "인천", "광주", "대전", "울산", "세종", "경기", "강원", "충북", "충남", "전북", "전남", "경북", "경남", "제주"]

results = []

def search_banks_in_region(region, page=1, size=15):
    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    params = {
        "query": f"{region} {KEYWORD}",
        "category_group_code": CATEGORY_GROUP_CODE,
        "page": page,
        "size": size
    }
    response = requests.get(url, headers=HEADERS, params=params)
    return response.json()

# 📌 모든 지역에 대해 반복 수집
for region in REGIONS:
    for page in range(1, 46):  # 카카오맵은 45페이지가 최대
        print(f"📍 {region} - {page}페이지 수집 중...")
        data = search_banks_in_region(region, page)
        
        if "documents" not in data or not data["documents"]:
            print(f"✅ {region} 지역 수집 완료.")
            break
        
        for doc in data["documents"]:
            name = doc["place_name"]
            lat = doc["y"]
            lon = doc["x"]
            results.append([name, lat, lon])
        
        time.sleep(0.3)  # API 요청 제한 보호

# 📌 데이터프레임 생성
df = pd.DataFrame(results, columns=["이름", "위도", "경도"])

# 📌 중복 제거 (지점명 + 좌표 기준)
df.drop_duplicates(inplace=True)

# 📌 결과 출력
# 📌 CSV 파일로 저장
output_file = "국민-2.csv"  # 원하는 파일 이름 설정
df.to_csv(output_file, index=False, encoding="utf-8-sig")  # 인덱스 없이 저장
print(f"✅ 완료! 저장된 파일: {output_file}")
```

```python
import requests
import pandas as pd
import time

# 카카오 API 키 설정
KAKAO_API_KEY = 'f32e9311c06e941799af2d38345741de'  # 여기에 본인의 카카오 API 키를 입력하세요
HEADERS = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}

# 시군구 중심 좌표 리스트 (예시로 일부만 포함, 전체 리스트는 추가 필요)
regions = [
    {"region": "서울특별시 강남구", "lat": 37.5172, "lon": 127.0473},
    {"region": "부산광역시 해운대구", "lat": 35.1632, "lon": 129.1636},
    {"region": "대구광역시 중구", "lat": 35.8700, "lon": 128.6060},
    {"region": "인천광역시 남동구", "lat": 37.4470, "lon": 126.7317},
    {"region": "광주광역시 서구", "lat": 35.1547, "lon": 126.8530},
    {"region": "대전광역시 유성구", "lat": 36.3622, "lon": 127.3568},
    {"region": "울산광역시 남구", "lat": 35.5384, "lon": 129.3114},
    {"region": "세종특별자치시", "lat": 36.4800, "lon": 127.2890},
    {"region": "경기도 수원시", "lat": 37.2636, "lon": 127.0286},
    {"region": "강원도 춘천시", "lat": 37.8813, "lon": 127.7298},
    {"region": "충청북도 청주시", "lat": 36.6424, "lon": 127.4890},
    {"region": "충청남도 천안시", "lat": 36.8151, "lon": 127.1139},
    {"region": "전라북도 전주시", "lat": 35.8210, "lon": 127.1480},
    {"region": "전라남도 목포시", "lat": 34.8118, "lon": 126.3922},
    {"region": "경상북도 포항시", "lat": 36.0190, "lon": 129.3435},
    {"region": "경상남도 창원시", "lat": 35.2285, "lon": 128.6811},
    {"region": "제주특별자치도 제주시", "lat": 33.4996, "lon": 126.5312},
    # 추가적인 시군구 중심 좌표를 여기에 추가하세요
]

def search_bank(lat, lon, query="부산은행", radius=20000):
    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    results = []
    for page in range(1, 46):  # 최대 45페이지
        params = {
            "query": query,
            "x": lon,
            "y": lat,
            "radius": radius,
            "page": page
        }
        res = requests.get(url, headers=HEADERS, params=params).json()
        if 'documents' not in res or not res['documents']:
            break
        results.extend(res['documents'])
        if res.get("meta", {}).get("is_end", True):
            break
        time.sleep(0.2)  # API 호출 간 간격을 두어 과도한 요청 방지
    return results

all_results = []
for r in regions:
    print(f"{r['region']} 검색 중...")
    results = search_bank(r["lat"], r["lon"])
    all_results.extend(results)
    time.sleep(0.5)  # 지역 간 요청 간격

# 중복 제거 (place_url 기준)
df = pd.DataFrame(all_results)
df.drop_duplicates(subset="place_url", inplace=True)

# 필요한 열만 추출
df_result = df[['place_name', 'y', 'x']]
df_result.columns = ['은행이름', '위도', '경도']

# 결과 저장
df_result.to_csv("부산_Bank.csv", index=False, encoding="utf-8-sig")
```
